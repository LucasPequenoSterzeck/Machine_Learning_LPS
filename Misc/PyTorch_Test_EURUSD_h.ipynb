```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from sklearn.model_selection import train_test_split
from torch.utils.data import DataLoader, TensorDataset

# Dados de treinamento
X = df[60000:].drop(columns='STATUS')
y = df[60000:].STATUS

# Dividir os dados em treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05, random_state=42)

# Converter os dados para tensores do PyTorch
X_train = torch.tensor(X_train.values, dtype=torch.float32)
y_train = torch.tensor(y_train.values, dtype=torch.float32)
X_test = torch.tensor(X_test.values, dtype=torch.float32)

# Criar DataLoader para os dados de treinamento em batches
train_dataset = TensorDataset(X_train, y_train)
train_loader = DataLoader(train_dataset, batch_size=128, shuffle=True)

# Definir a arquitetura da rede neural
class NeuralNetwork(nn.Module):
    def __init__(self):
        super(NeuralNetwork, self).__init__()
        self.fc1 = nn.Linear(114, 64)
        self.fc3 = nn.Linear(64, 248)
        self.fc4 = nn.Linear(248, 248)
        self.fc5 = nn.Linear(248, 114)
        self.fc6 = nn.Linear(114, 64)
        self.fc7 = nn.Linear(64, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.fc1(x)
        x = torch.relu(x)
        x = self.fc3(x)
        x = torch.relu(x)
        x = self.fc4(x)
        x = torch.relu(x)
        x = self.fc5(x)
        x = torch.relu(x)
        x = self.fc6(x)
        x = torch.relu(x)
        x = self.fc7(x)
        x = self.sigmoid(x)
        return x

# Instanciar a rede neural
model = NeuralNetwork()

# Definir função de perda e otimizador
criterion = nn.BCELoss()
optimizer = optim.SGD(model.parameters(), lr=0.005)

# Treinamento da rede neural
num_epochs = 10000
for epoch in range(num_epochs):
    running_loss = 0.0
    for i, (inputs, labels) in enumerate(train_loader):
        # Forward pass
        outputs = model(inputs)
        loss = criterion(outputs.squeeze(), labels)

        # Backward pass e otimização
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    # Printando a loss média a cada 10 batches
    if (epoch+1) % 10 == 0:
        average_loss = running_loss / len(train_loader)
        print(f'Epoch {epoch+1}/{num_epochs}, Loss: {average_loss}')
```

```python

# Previsões
with torch.no_grad():
    predictions = model(X_test)
    predicted_labels = (predictions >= 0.5).squeeze().int()

# Exibindo as previsões
for pred in predicted_labels:
    print("Evento" if pred == 1 else "Não evento")
```
