```python
import pandas as pd

data = {
    'Unidade': ['Piracicaba', 'Piracicaba', 'Charqueada', 'Charqueada', 'Piracicaba', 'Piracicaba', 'Charqueada', 'Charqueada', 'Piracicaba', 'Piracicaba', 'Charqueada'],
    'Moagem': [25000, 35000, 15000, 10000, 45000, 20000, 8000, 12000, 30000, 40000, 20000],
    'Momento': ['10/07/2023 05:00:00', '10/07/2023 06:00:00', '10/07/2023 05:00:00', '10/07/2023 06:00:00', '10/07/2023 07:00:00', '10/07/2023 08:00:00', '10/07/2023 07:00:00', '10/07/2023 08:00:00', '10/07/2023 09:00:00', '10/07/2023 10:00:00', '10/07/2023 11:00:00'],
    'Apt_Produtivo': [2000, 2800, 1200, 800, 3200, 1600, 600, 1000, 2400, 3200, 1600],
    'Apt_Auxiliar': [1500, 2000, 900, 600, 2400, 1200, 400, 800, 1800, 2400, 1200],
    'Apt_Manutencao': [400, 300, 600, 700, 200, 500, 800, 700, 300, 200, 500],
    'Apt_Perdido': [700, 500, 300, 500, 400, 700, 500, 400, 500, 400, 700]
}

df = pd.DataFrame(data)

# Exibindo o DataFrame
df
```

```python
# Criando variáveis
import torch
import numpy as np

inputs = torch.from_numpy(df.drop(columns=['Moagem','Unidade','Momento']).values).float()
target = torch.from_numpy(df.Moagem.values).float()

pesos = torch.rand([4], requires_grad=True, dtype=torch.float)
bias = torch.rand([1], requires_grad=True, dtype=torch.float)

# Declarando modelo
def modelo(entrada):
  return entrada @ pesos.t() + bias

predicao = modelo(inputs)

# Função de custo
def custo(predict, target):
  diff = (predict - target)
  return torch.sum(diff * diff) / diff.numel()

loss = custo(predicao, target)
loss.backward()
# Vendo gradientes:
print('Antigos pesos e bias:')
print(pesos)
print(bias)
print('Gradientes:')
print(pesos.grad)
print(bias.grad)
with torch.no_grad():
  pesos -= pesos.grad * 0.00001 # 1e-5
  bias -= bias.grad * 0.00001 # 1e-5
  pesos.grad.zero_()
  bias.grad.zero_()
print('Novos pesos e bias:')
print(pesos)
print(bias)
print(f'--> Estamos saindo de um erro de: {loss}, para...')
```

```python
epochs = 100
for i in range(epochs):
    preds = modelo(inputs)
    loss = custo(preds, target)
    loss.backward()
    with torch.no_grad():
        pesos -= pesos.grad * 1e-7
        bias -= bias.grad * 1e-7
        pesos.grad.zero_()
        bias.grad.zero_()
    if (i+1) % 10 == 0:
      print(f'Epoch [{i+1}/{epochs}], Loss: {loss}')
```

```python
print('Comparando agora caso a caso, temos:')
for prev, real in zip(preds, target):
  print(f'Previsto {round(prev.item())} |vs| Real {round(real.item())}  [Diff.: {round(prev.item()-real.item())} - {round((prev.item()-real.item())/real.item()*100,2)}%]')
```

# Agora aplicando modelo segmentado por unidade

```python
# Dicionario para conter modelos: NOME = (pesos, bias)
g_model = {}

# iniciando modelos:
for unidade in df.Unidade.unique():
  pesos = torch.rand([4], requires_grad=True, dtype=torch.float)
  bias = torch.rand([1], requires_grad=True, dtype=torch.float)
  g_model[unidade] = [pesos,bias]
g_model
```

```python
# Declarando modelo
def modelo(entrada, pesos, bias):
  return entrada @ pesos.t() + bias

# Função de custo
def custo(predict, target):
  diff = (predict - target)
  return torch.sum(diff * diff) / diff.numel()

# Taxa de aprendizado:
learning_rate = 0.000000004

epochs = 10000
for unidade in df.Unidade.unique():
  print(f'\nIniciando treinamento de modelo da unidade {unidade}')
  input_unidade = torch.from_numpy(df[df.Unidade==unidade].drop(columns=['Moagem','Unidade','Momento']).values).float()
  target_unidade = torch.from_numpy(df[df.Unidade==unidade].Moagem.values).float()
  for i in range(epochs):
      preds = modelo(input_unidade, g_model[unidade][0], g_model[unidade][1])
      loss = custo(preds, target_unidade)
      loss.backward()
      with torch.no_grad():
          g_model[unidade][0] -= g_model[unidade][0].grad * learning_rate
          g_model[unidade][1] -= g_model[unidade][1].grad * learning_rate
          g_model[unidade][0].grad.zero_()
          g_model[unidade][1].grad.zero_()
      if (i+1) % 1000 == 0:
        print(f'Epoch [{i+1}/{epochs}], Loss: {loss}')
```

## Comparando previsões para Unidades:

```python
for unidade in df.Unidade.unique():
  print(f'\n[INFO] Comparando previsões Unidade: {unidade}.')
  input_unidade = torch.from_numpy(df[df.Unidade==unidade].drop(columns=['Moagem','Unidade','Momento']).values).float()
  target_unidade = torch.from_numpy(df[df.Unidade==unidade].Moagem.values).float()
  preds = modelo(input_unidade, g_model[unidade][0], g_model[unidade][1])
  for prev, real in zip(preds, target):
    print(f'Previsto {round(prev.item())} |vs| Real {round(real.item())}  [Diff.: {round(prev.item()-real.item())} - {round((prev.item()-real.item())/real.item()*100,2)}%]')
```
